Sample dataset

Untuk menghasilkan data area kita dapat menggunakan fungsi np.meshgrid().
Nilai yang ditampung X, Y, dan Z adalah array 2 dimensi.

```python
n = np.linspace(-3.0, 3.0, 100) # (Start, End, Amount)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
(100, 100)
(100, 100)
(100, 100)
```

3D Wireframe

Pertama buat objek figure dan objek axes.
Namun untuk membuat axes 3D, kita perlu menyertakan parameter projection yang kita beri nilai '3d'.
Untuk melakukan plotting dalam wireframe plot kita dapat memanggil fungsi ax.plot_wireframe().

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

3D Surface Plot

Berbeda dengan 3D Wireframe plot. 3D Surface Plot akan memenuhi areanya dengan warna.
Letak perbedaan fungsi ada di ax.plot_surface() sedangkan sebelumnya kita menggunakan ax.plot_wireframe().

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# ax.plot_surface(x, y, z, color='orange')
# ax.plot_surface(x, y, z, color='pink')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```
